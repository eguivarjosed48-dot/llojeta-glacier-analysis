Definir rutas base y Verificar

```python
from pathlib import Path
import sys

REPO_PATH = Path("/content/drive/MyDrive/geo_projects/llojeta-glacier-analysis/repo")
CONFIG_PATH = REPO_PATH / "config"
SRC_PATH = REPO_PATH / "src"
PROJECTIO_PATH = SRC_PATH / "projectio"

print("REPO_PATH     =", REPO_PATH)
print("CONFIG_PATH   =", CONFIG_PATH)
print("SRC_PATH      =", SRC_PATH)
print("PROJECTIO_PATH=", PROJECTIO_PATH)

assert REPO_PATH.exists(), f"No existe: {REPO_PATH}"
assert CONFIG_PATH.exists(), f"No existe: {CONFIG_PATH}"
assert SRC_PATH.exists(), f"No existe: {SRC_PATH}"
assert PROJECTIO_PATH.exists(), f"No existe: {PROJECTIO_PATH}"
```

crear config/paths.yaml

```python
paths_yaml = """project_root: /content/drive/MyDrive/geo_projects/llojeta-glacier-analysis/repo

data:
  raw: data/raw
  interim: data/interim
  processed: data/processed
  publish: data/publish

outputs:
  figures: outputs/figures
  tables: outputs/tables
  reports: outputs/reports
  qa: outputs/qa

web:
  root: web
  assets: web/assets
  maps: web/maps
  scenes: web/scenes
"""

paths_file = CONFIG_PATH / "paths.yaml"
paths_file.write_text(paths_yaml, encoding="utf-8")

print("Archivo creado:", paths_file)
print()
print(paths_file.read_text(encoding="utf-8"))
```

crear módulo src/projectio/paths.py

```python
paths_code = """
from pathlib import Path
import yaml


class Paths:
    def __init__(self, repo_path):
        self.repo_path = Path(repo_path)
        config_file = self.repo_path / "config" / "paths.yaml"

        if not config_file.exists():
            raise FileNotFoundError(f"No existe: {config_file}")

        with open(config_file, "r", encoding="utf-8") as f:
            cfg = yaml.safe_load(f)

        self.root = Path(cfg["project_root"])
        self.cfg = cfg

    def raw(self):
        return self.root / self.cfg["data"]["raw"]

    def interim(self):
        return self.root / self.cfg["data"]["interim"]

    def processed(self):
        return self.root / self.cfg["data"]["processed"]

    def publish(self):
        return self.root / self.cfg["data"]["publish"]

    def figures(self):
        return self.root / self.cfg["outputs"]["figures"]

    def tables(self):
        return self.root / self.cfg["outputs"]["tables"]

    def reports(self):
        return self.root / self.cfg["outputs"]["reports"]

    def qa(self):
        return self.root / self.cfg["outputs"]["qa"]

    def web_root(self):
        return self.root / self.cfg["web"]["root"]

    def web_assets(self):
        return self.root / self.cfg["web"]["assets"]

    def web_maps(self):
        return self.root / self.cfg["web"]["maps"]

    def web_scenes(self):
        return self.root / self.cfg["web"]["scenes"]
"""

paths_py_file = PROJECTIO_PATH / "paths.py"
paths_py_file.write_text(paths_code, encoding="utf-8")

print("Archivo creado:", paths_py_file)
```

importar el modulo

```python
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

from projectio.paths import Paths

print("Import correcto")
```

probar uso

```python
cfg = Paths(REPO_PATH)

print("RAW       :", cfg.raw())
print("INTERIM   :", cfg.interim())
print("PROCESSED :", cfg.processed())
print("PUBLISH   :", cfg.publish())
print("FIGURES   :", cfg.figures())
print("TABLES    :", cfg.tables())
print("REPORTS   :", cfg.reports())
print("QA        :", cfg.qa())
print("WEB ROOT  :", cfg.web_root())
print("WEB ASSETS:", cfg.web_assets())
print("WEB MAPS  :", cfg.web_maps())
print("WEB SCENES:", cfg.web_scenes())
```

asegurar directorios por si faltan

```python
dirs_to_check = [
    cfg.raw(),
    cfg.interim(),
    cfg.processed(),
    cfg.publish(),
    cfg.figures(),
    cfg.tables(),
    cfg.reports(),
    cfg.qa(),
    cfg.web_root(),
    cfg.web_assets(),
    cfg.web_maps(),
    cfg.web_scenes(),
]

for d in dirs_to_check:
    d.mkdir(parents=True, exist_ok=True)
    print("OK:", d)
```
