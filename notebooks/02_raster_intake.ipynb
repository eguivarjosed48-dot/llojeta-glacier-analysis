```python
from pathlib import Path
import sys

REPO_PATH = Path("/content/drive/MyDrive/geo_projects/llojeta-glacier-analysis/repo")
SRC_PATH = REPO_PATH / "src"

assert REPO_PATH.exists(), f"No existe: {REPO_PATH}"
assert SRC_PATH.exists(), f"No existe: {SRC_PATH}"

if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

print("REPO_PATH:", REPO_PATH)
print("SRC_PATH :", SRC_PATH)
```

Imports

```python
import json
import shutil
import pandas as pd

from projectio.paths import Paths
from projectio.raster_inventory import (
    build_raster_inventory,
    add_basic_qa,
    summarize_pairs,
)
```

Rutas proyecto y carpeta fuente

```python
cfg = Paths(REPO_PATH)

SOURCE_DIR = Path("/content/drive/MyDrive/geo_projects/llojeta-glacier-analysis/Datos_Originales")
EXPECTED_EPOCHS = ["2017", "2018", "2020", "2023", "2025_A", "2025_B"]

assert SOURCE_DIR.exists(), f"No existe SOURCE_DIR: {SOURCE_DIR}"

QA_DIR = cfg.qa()
QA_DIR.mkdir(parents=True, exist_ok=True)

RAW_BASE = cfg.raw() / "llojeta"
RAW_BASE.mkdir(parents=True, exist_ok=True)

print("SOURCE_DIR:", SOURCE_DIR)
print("QA_DIR    :", QA_DIR)
print("RAW_BASE  :", RAW_BASE)
```

# 5. inventario de la fuente original

```python
source_inventory_df = build_raster_inventory(SOURCE_DIR)
source_inventory_df = add_basic_qa(source_inventory_df)

print("=== INVENTARIO FUENTE ORIGINAL ===")
print("Total TIFF:", len(source_inventory_df))
print("QA ok     :", int(source_inventory_df["qa_ok"].sum()))
print("QA fail   :", int((~source_inventory_df["qa_ok"]).sum()))

source_inventory_df[[
    "filename", "epoch", "product_type", "crs", "width", "height",
    "count", "dtype", "nodata", "res_x", "res_y", "qa_ok", "error"
]]
```

6. Copia normalizada a data/raw/llojeta

```python
copied = []

valid_rows = source_inventory_df[source_inventory_df["qa_ok"] == True].copy()

for _, row in valid_rows.iterrows():
    epoch = row["epoch"]
    product_type = row["product_type"].lower()
    src = Path(row["source_path"])

    dst_dir = RAW_BASE / epoch / product_type
    dst_dir.mkdir(parents=True, exist_ok=True)

    dst = dst_dir / src.name
    shutil.copy2(src, dst)

    copied.append({
        "epoch": epoch,
        "product_type": product_type,
        "source_original_path": str(src),
        "project_raw_path": str(dst),
    })

print(f"Copiados {len(copied)} archivos al árbol operativo del proyecto")
```

7. guardar log de copia

```python
copy_log_path = QA_DIR / "copy_log.json"

with open(copy_log_path, "w", encoding="utf-8") as f:
    json.dump(copied, f, indent=2, ensure_ascii=False)

print("Log de copia guardado en:", copy_log_path)
```

8. Listado Tiff normalizados en data/raw/llojeta

```python
raw_tif_files = sorted(RAW_BASE.rglob("*.tif"))

print(f"Se encontraron {len(raw_tif_files)} TIFF en data/raw/llojeta:\n")
for p in raw_tif_files:
    print("-", p.relative_to(RAW_BASE))
```

9. Inventario oficial del proyecto desde data/raw/llojeta

```python
records = []

for tif in raw_tif_files:
    epoch = tif.parent.parent.name
    product_type = tif.parent.name.lower()

    records.append({
        "filename": tif.name,
        "epoch": epoch,
        "product_type": product_type,
        "source_path": str(tif),
    })

raw_inventory_df = pd.DataFrame(records)
raw_inventory_df
```

10. Inventario enriquecido con modulo QA

```python
from projectio.raster_inventory import inspect_raster

meta_records = []

for _, row in raw_inventory_df.iterrows():
    meta = inspect_raster(Path(row["source_path"]))
    meta_records.append(meta)

meta_df = pd.DataFrame(meta_records)

project_inventory_df = pd.concat(
    [raw_inventory_df.reset_index(drop=True), meta_df.reset_index(drop=True)],
    axis=1
)

project_inventory_df["valid_name"] = True
project_inventory_df = add_basic_qa(project_inventory_df)

print("=== INVENTARIO OFICIAL DEL PROYECTO ===")
print("Total TIFF:", len(project_inventory_df))
print("QA ok     :", int(project_inventory_df["qa_ok"].sum()))
print("QA fail   :", int((~project_inventory_df["qa_ok"]).sum()))

project_inventory_df[[
    "filename", "epoch", "product_type", "crs", "width", "height",
    "count", "dtype", "nodata", "res_x", "res_y", "qa_ok", "error"
]]
```

11. Validacion pares DSM/Ortho por Epoch

```python
pairs_df = summarize_pairs(project_inventory_df, expected_epochs=EXPECTED_EPOCHS)
pairs_df
```

12. Inventario oficial y resumen de pares

```python
project_inventory_csv = QA_DIR / "project_raster_inventory.csv"
project_inventory_json = QA_DIR / "project_raster_inventory.json"
pairs_csv = QA_DIR / "project_raster_pairs_summary.csv"

project_inventory_df.to_csv(project_inventory_csv, index=False, encoding="utf-8")

with open(project_inventory_json, "w", encoding="utf-8") as f:
    json.dump(project_inventory_df.to_dict(orient="records"), f, indent=2, ensure_ascii=False)

pairs_df.to_csv(pairs_csv, index=False, encoding="utf-8")

print("Guardado inventario oficial CSV :", project_inventory_csv)
print("Guardado inventario oficial JSON:", project_inventory_json)
print("Guardado resumen de pares       :", pairs_csv)
```

13. Fallos QA oficiales

```python
failed_df = project_inventory_df.loc[~project_inventory_df["qa_ok"]].copy()

print("Número de fallos QA oficiales:", len(failed_df))
failed_df[[
    "filename",
    "epoch",
    "product_type",
    "error",
    "crs",
    "width",
    "height",
    "count",
]]
```

14. Resumen Final

```python
print("=== RESUMEN FINAL ===")
print("Fuente original                :", SOURCE_DIR)
print("Árbol operativo del proyecto   :", RAW_BASE)
print("Total TIFF copiados            :", len(copied))
print("Total TIFF inventario oficial  :", len(project_inventory_df))
print("QA ok                          :", int(project_inventory_df["qa_ok"].sum()))
print("QA fail                        :", int((~project_inventory_df["qa_ok"]).sum()))

print("\nPares oficiales por epoch:")
display(pairs_df)

print("\nArchivos con fallo QA oficial:")
display(failed_df)
```
