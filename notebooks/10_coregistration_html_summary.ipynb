```python
import json
import pandas as pd
import numpy as np
```

## Configuración

```python
BASE_EPOCH = '2025_B'

DIAG_DIR = REPO_PATH / 'outputs' / 'coregistration_diagnostics'
WEB_DIR = REPO_PATH / 'web'
WEB_TABLES_DIR = WEB_DIR / 'tables'
FINAL_DIR = REPO_PATH / 'outputs' / 'coregistration_final'

WEB_TABLES_DIR.mkdir(parents=True, exist_ok=True)
FINAL_DIR.mkdir(parents=True, exist_ok=True)

scenario_csv = DIAG_DIR / 'coreg_diag_scenario_comparison.csv'
extended_csv = DIAG_DIR / 'coreg_diag_extended_stats.csv'
decision_csv = DIAG_DIR / 'coreg_diag_decision_table.csv'

assert scenario_csv.exists(), f'No existe: {scenario_csv}'
assert extended_csv.exists(), f'No existe: {extended_csv}'

print('scenario_csv:', scenario_csv)
print('extended_csv:', extended_csv)
print('decision_csv:', decision_csv if decision_csv.exists() else 'No existe aún')
```

## Carga de datos

```python
scenario_df = pd.read_csv(scenario_csv)
extended_df = pd.read_csv(extended_csv)
decision_df = pd.read_csv(decision_csv) if decision_csv.exists() else None

print('scenario_df:', len(scenario_df))
print('extended_df:', len(extended_df))
print('decision_df:', len(decision_df) if decision_df is not None else 0)

display(scenario_df.head())
display(extended_df.head())
if decision_df is not None:
    display(decision_df.head())
```

## Helpers

```python
def choose_best_scenario_for_epoch(sdf, epoch):
    sub = sdf[sdf['target_epoch'].astype(str) == str(epoch)].copy()
    raw_sub = sub[sub['scenario'].astype(str).str.endswith('_raw')].copy()
    if raw_sub.empty:
        raise ValueError(f'No hay escenarios raw para epoch={epoch}')
    raw_sub = raw_sub.sort_values(
        by=['nmad_after', 'median_abs_after', 'mean_abs_after', 'std_after'],
        ascending=[True, True, True, True]
    ).reset_index(drop=True)
    return raw_sub.iloc[0]

def classify_final_decision(row):
    median_improved = bool(row['median_improved'])
    nmad_improved = bool(row['nmad_improved'])
    std_improved = bool(row['std_improved'])
    mean_improved = bool(row['mean_improved'])
    median_after = float(row['median_after'])
    mean_after = float(row['mean_after'])

    if (not median_improved) and (not mean_improved):
        return 'use_original'
    if median_improved and nmad_improved and std_improved and abs(mean_after) <= 3.0:
        return 'use_corrected'
    if median_improved and nmad_improved and abs(mean_after) <= 5.0:
        return 'use_corrected'
    if median_improved and abs(median_after) <= 0.25 and abs(mean_after) <= 2.0:
        return 'use_corrected_nearly_neutral'
    return 'review'

def rationale_from_row(row):
    decision = row['final_decision']
    if decision == 'use_corrected':
        return 'La corrección mejora el ajuste central y la dispersión robusta sobre terreno estable.'
    if decision == 'use_corrected_nearly_neutral':
        return 'La corrección tiene efecto pequeño, pero no degrada de forma relevante el ajuste.'
    if decision == 'use_original':
        return 'La corrección empeora el ajuste central; conviene conservar el DEM original.'
    return 'La corrección muestra señales mixtas y requiere revisión adicional.'

def decision_label(decision):
    return {
        'use_corrected': 'Use corrected',
        'use_corrected_nearly_neutral': 'Use corrected (near neutral)',
        'use_original': 'Use original',
        'review': 'Review',
    }.get(decision, decision)

def badge_class(decision):
    return {
        'use_corrected': 'badge badge-green',
        'use_corrected_nearly_neutral': 'badge badge-blue',
        'use_original': 'badge badge-red',
        'review': 'badge badge-amber',
    }.get(decision, 'badge')

def fmt(x, nd=3):
    if pd.isna(x):
        return ''
    return f'{float(x):.{nd}f}'
```

## Selección final por epoch

```python
epochs = sorted(scenario_df['target_epoch'].astype(str).unique().tolist())

records = []
for epoch in epochs:
    best = choose_best_scenario_for_epoch(scenario_df, epoch).to_dict()
    best['final_decision'] = classify_final_decision(best)
    best['decision_label'] = decision_label(best['final_decision'])
    best['decision_rationale'] = rationale_from_row(best)
    best['publish'] = best['final_decision'] in ['use_corrected', 'use_corrected_nearly_neutral', 'use_original']
    best['selected_product'] = (
        'corrected' if best['final_decision'] in ['use_corrected', 'use_corrected_nearly_neutral'] else
        'original' if best['final_decision'] == 'use_original' else
        'review'
    )
    records.append(best)

final_df = pd.DataFrame(records).sort_values('target_epoch').reset_index(drop=True)
display(final_df)
```

## Tabla compacta para publicación

```python
publish_df = final_df.copy()

publish_table_df = publish_df[[
    'target_epoch',
    'scenario',
    'mean_after',
    'median_after',
    'std_after',
    'nmad_after',
    'mean_improved',
    'median_improved',
    'std_improved',
    'nmad_improved',
    'decision_label',
    'decision_rationale',
    'selected_product',
    'publish',
]].copy()

publish_table_df = publish_table_df.rename(columns={
    'target_epoch': 'Epoch',
    'scenario': 'Scenario',
    'mean_after': 'Mean after',
    'median_after': 'Median after',
    'std_after': 'STD after',
    'nmad_after': 'NMAD after',
    'mean_improved': 'Mean improved',
    'median_improved': 'Median improved',
    'std_improved': 'STD improved',
    'nmad_improved': 'NMAD improved',
    'decision_label': 'Decision',
    'decision_rationale': 'Rationale',
    'selected_product': 'Selected product',
    'publish': 'Publish',
})

display(publish_table_df)
```

## HTML listo para GitHub Pages

```python
html_df = final_df.copy()

html_df['Mean after'] = html_df['mean_after'].map(lambda x: fmt(x, 3))
html_df['Median after'] = html_df['median_after'].map(lambda x: fmt(x, 3))
html_df['STD after'] = html_df['std_after'].map(lambda x: fmt(x, 3))
html_df['NMAD after'] = html_df['nmad_after'].map(lambda x: fmt(x, 3))
html_df['Scenario'] = html_df['scenario']
html_df['Decision'] = html_df.apply(
    lambda r: '<span class="%s">%s</span>' % (badge_class(r['final_decision']), decision_label(r['final_decision'])),
    axis=1
)
html_df['Selected product'] = html_df['selected_product']
html_df['Rationale'] = html_df['decision_rationale']

html_table_df = html_df[[
    'target_epoch', 'Scenario', 'Mean after', 'Median after', 'STD after',
    'NMAD after', 'Decision', 'Selected product', 'Rationale'
]].rename(columns={'target_epoch': 'Epoch'})

table_html = html_table_df.to_html(index=False, escape=False, classes=['coreg-table'])

style_block = '''<style>
body { font-family: Arial, Helvetica, sans-serif; margin: 24px; color: #1f2937; background: #ffffff; }
h1 { margin-bottom: 8px; }
p.meta { color: #4b5563; margin-top: 0; }
table.coreg-table { border-collapse: collapse; width: 100%; font-size: 14px; }
.coreg-table th, .coreg-table td { border: 1px solid #e5e7eb; padding: 10px 12px; vertical-align: top; text-align: left; }
.coreg-table th { background: #f3f4f6; }
.coreg-table tr:nth-child(even) { background: #fafafa; }
.badge { display: inline-block; border-radius: 999px; padding: 4px 10px; font-size: 12px; font-weight: 700; white-space: nowrap; }
.badge-green { background: #dcfce7; color: #166534; }
.badge-blue { background: #dbeafe; color: #1d4ed8; }
.badge-amber { background: #fef3c7; color: #92400e; }
.badge-red { background: #fee2e2; color: #991b1b; }
code { background: #f3f4f6; padding: 2px 6px; border-radius: 4px; }
</style>'''

full_html = '<!DOCTYPE html><html lang="en"><head><meta charset="utf-8"><meta name="viewport" content="width=device-width, initial-scale=1"><title>Coregistration Summary</title>' + style_block + '</head><body><h1>Coregistration Summary Table</h1><p class="meta">Base epoch: <code>' + BASE_EPOCH + '</code></p>' + table_html + '</body></html>'

snippet_html = table_html

display(html_table_df)
print(full_html[:1200])
```

## Exportación

```python
summary_csv = FINAL_DIR / 'coregistration_summary_table.csv'
summary_json = FINAL_DIR / 'coregistration_summary_table.json'
html_full_path = WEB_TABLES_DIR / 'coregistration_summary.html'
html_snippet_path = WEB_TABLES_DIR / 'coregistration_summary_snippet.html'

publish_table_df.to_csv(summary_csv, index=False, encoding='utf-8')
final_df.to_json(summary_json, orient='records', force_ascii=False, indent=2)

with open(html_full_path, 'w', encoding='utf-8') as f:
    f.write(full_html)

with open(html_snippet_path, 'w', encoding='utf-8') as f:
    f.write(snippet_html)

print('Guardado:', summary_csv)
print('Guardado:', summary_json)
print('Guardado:', html_full_path)
print('Guardado:', html_snippet_path)
```

## Cómo usar en GitHub Pages

- Usa `web/tables/coregistration_summary.html` como página lista para publicar.
- Usa `web/tables/coregistration_summary_snippet.html` si quieres incrustar solo la tabla dentro de otra página.
- Si quieres un criterio más conservador o más laxo, modifica únicamente `classify_final_decision()`.
