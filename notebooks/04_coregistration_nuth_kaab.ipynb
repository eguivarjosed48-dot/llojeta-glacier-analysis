```python
import pandas as pd
import rasterio

from projectio.paths import Paths
from processing.coregistration_nk import coregister_dem_nk_deramp
```

```python
cfg = Paths(REPO_PATH)

QA_DIR = cfg.qa()
inventory_csv = QA_DIR / "project_raster_inventory.csv"

assert inventory_csv.exists(), f"No existe inventario oficial: {inventory_csv}"

inventory_df = pd.read_csv(inventory_csv)

print("Inventario cargado:", len(inventory_df))
inventory_df.head()
```

```python
BASE_EPOCH = "2025_B"
TARGET_EPOCHS = ["2017", "2018", "2020", "2023", "2025_A"]      # luego puedes poner varias: ["2017", "2018", "2020", "2023", "2025_A"]
FACTOR_GIRO = -0.05
POLY_ORDER = 1

MASK_PATH = Path("/content/drive/MyDrive/geo_projects/llojeta-glacier-analysis/Datos_Originales/mascaras/poligono_roca_estable/poligono_roca_estable.shp")
OUTPUT_DIR = REPO_PATH / "outputs" / "coregistration_nk"

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("BASE_EPOCH   :", BASE_EPOCH)
print("TARGET_EPOCHS:", TARGET_EPOCHS)
print("FACTOR_GIRO  :", FACTOR_GIRO)
print("POLY_ORDER   :", POLY_ORDER)
print("MASK_PATH    :", MASK_PATH)
print("OUTPUT_DIR   :", OUTPUT_DIR)
print("MASK exists  :", MASK_PATH.exists())
```

```python
mask_base = MASK_PATH.with_suffix("")
for ext in [".shp", ".shx", ".dbf", ".prj"]:
    p = mask_base.with_suffix(ext)
    print(ext, p.exists(), p)
```

```python
def get_dsm_path_for_epoch(inventory_df, epoch):
    row = inventory_df[
        (inventory_df["epoch"] == epoch) &
        (inventory_df["product_type"] == "dsm") &
        (inventory_df["qa_ok"] == True)
    ]
    assert len(row) == 1, f"No se encontró un único DSM válido para epoch={epoch}"
    return Path(row.iloc[0]["source_path"])
```

```python
base_dem_path = get_dsm_path_for_epoch(inventory_df, BASE_EPOCH)
print("Base DEM:", base_dem_path)
print("Existe  :", base_dem_path.exists())
```

```python
target_preview = []

for epoch in TARGET_EPOCHS:
    target_dem_path = get_dsm_path_for_epoch(inventory_df, epoch)
    target_preview.append({
        "epoch": epoch,
        "target_dem_path": str(target_dem_path),
        "exists": target_dem_path.exists(),
    })

target_preview_df = pd.DataFrame(target_preview)
target_preview_df
```

```python
results = []

for epoch in TARGET_EPOCHS:
    target_dem_path = get_dsm_path_for_epoch(inventory_df, epoch)
    output_path = OUTPUT_DIR / f"{epoch}_DSM_nk_corrected.tif"

    print(f"\n[{epoch}] ---------------------------------")
    print("Base   :", base_dem_path)
    print("Target :", target_dem_path)
    print("Output :", output_path)

    try:
        result = coregister_dem_nk_deramp(
            base_dem_path=base_dem_path,
            target_dem_path=target_dem_path,
            mask_vector_path=MASK_PATH,
            output_path=output_path,
            factor_giro=FACTOR_GIRO,
            poly_order=POLY_ORDER,
        )

        result["epoch"] = epoch
        result["base_epoch"] = BASE_EPOCH
        results.append(result)

        print("✔ OK")

    except Exception as e:
        results.append({
            "epoch": epoch,
            "base_epoch": BASE_EPOCH,
            "base_dem_path": str(base_dem_path),
            "target_dem_path": str(target_dem_path),
            "mask_vector_path": str(MASK_PATH),
            "output_path": str(output_path),
            "factor_giro": FACTOR_GIRO,
            "poly_order": POLY_ORDER,
            "status": "error",
            "error": str(e),
        })

        print("❌ ERROR:", e)
```

```python
results_df = pd.DataFrame(results)
results_df
```

```python
results_csv = OUTPUT_DIR / "coregistration_nk_results.csv"
results_df.to_csv(results_csv, index=False, encoding="utf-8")

print("Resultados guardados en:", results_csv)
```

```python
output_checks = []

for _, row in results_df.iterrows():
    out_path = Path(row["output_path"])
    output_checks.append({
        "epoch": row["epoch"],
        "status": row["status"],
        "output_exists": out_path.exists(),
        "output_size_bytes": out_path.stat().st_size if out_path.exists() else None,
        "output_path": str(out_path),
    })

output_checks_df = pd.DataFrame(output_checks)
output_checks_df
```

```python
inspection_records = []

ok_rows = results_df[results_df["status"] == "ok"].copy()

for _, row in ok_rows.iterrows():
    out_path = Path(row["output_path"])

    try:
        with rasterio.open(out_path) as src:
            inspection_records.append({
                "epoch": row["epoch"],
                "output_path": str(out_path),
                "crs": str(src.crs) if src.crs else None,
                "width": src.width,
                "height": src.height,
                "count": src.count,
                "dtype": src.dtypes[0] if src.count > 0 else None,
                "nodata": src.nodata,
                "res_x": src.res[0],
                "res_y": src.res[1],
            })
    except Exception as e:
        inspection_records.append({
            "epoch": row["epoch"],
            "output_path": str(out_path),
            "crs": None,
            "width": None,
            "height": None,
            "count": None,
            "dtype": None,
            "nodata": None,
            "res_x": None,
            "res_y": None,
            "error": str(e),
        })

inspection_df = pd.DataFrame(inspection_records)
inspection_df
```

```python
print("=== RESUMEN FINAL ===")
print("Base epoch       :", BASE_EPOCH)
print("Total targets    :", len(TARGET_EPOCHS))
print("OK               :", int((results_df["status"] == "ok").sum()))
print("ERROR            :", int((results_df["status"] == "error").sum()))
print("Resultados CSV   :", results_csv)
print("Output directory :", OUTPUT_DIR)

print("\nResultados:")
display(results_df)

print("\nChequeo de outputs:")
display(output_checks_df)

print("\nInspección raster de outputs OK:")
display(inspection_df)
```
